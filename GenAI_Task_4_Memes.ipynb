```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

# Load the image
img = Image.open(r"D:\Doraemon_pic.webp")

# Initialize ImageDraw
draw = ImageDraw.Draw(img)

# Load a font
font = ImageFont.truetype(r"C:\Users\hp\Downloads\Roboto (1)\Roboto-Bold.ttf", 25)

# Define the text for each meme
text_1 = """Job description: '5 years 
experience needed 
for entry-level.!"""
text_2 = """When the interviewer
asks if you know 'every 
tool' on the planet.!"""
text_3 = """Got an interview call 
after 100 applications.!"""
text_4 = """When you realize your 
first paycheck is real!"""

# Add text to each image
draw.text((50, 10), text=text_1, fill=(255, 0, 0), font=font, align="center")  # Top left image
draw.text((400, 10), text=text_2, fill=(0, 0, 255), font=font, align="center")  # Top right image
draw.text((50, 310), text=text_3, fill=(0, 255, 0), font=font, align="center")  # Bottom left image
draw.text((400, 310), text=text_4, fill=(255, 0, 0), font=font, align="center")  # Bottom right image

# Display the image
plt.imshow(img)
# plt.axis('off')  # Hide axes
plt.show()
```
